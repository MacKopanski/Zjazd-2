## Metody statyczne

```python
class Samochod:
    ILOSC_KOL = 4
    LITRY_NA_GALON = 3.78
    KILOMETROW_NA_MILE = 1.609
    def __init__(self, marka, model, kolor, przebieg, moc, spalanie):
        self.marka = marka
        self.model = model
        self.kolor = kolor
        self.przebieg = przebieg
        self.moc = moc
        self.spalanie = spalanie
    
    @property
    def pelna_nazwa(self):
        return "To jest {} {}".format(self.marka, self.model)
    
    # metody specjalne __str__
    def __str__(self):
        return self.pelna_nazwa
    
    def __repr__(self):
        return "Obiekt klasy: {}, marka: {}. model {}".format(
            self.__class__,
            self.marka,
            self.model
        )
    
    def __gt__(self, other):
        return self.przebieg > other.przebieg
    
    def wyswietl_przebieg(self):
        print(self.przebieg)
        
    @staticmethod
    def konwertuj_mimile_na_galon_na_lna100km (mile_na_galon):
        return (mile_na_galon* 100*Samochod.LITRY_NA_GALON / Samochod.KILOMETROW_NA_MILE)
```

## Podklasy
##### * Stworz klase Czlowiek,
    * Czlowiek moze jesc, pic, spac
    * Czlowiek ma parametry imie, nazwisko, wiek
    * Dodatkowo parametry czy_najeczony, czy_wyspany, czy_spragniony

* Stworz podklade CzlowiekNowoczesny
    * Czlowiek nowoczesny dziedzizczy po Czlowiek,
    * Dodatkowo ma:
        * Konto na github
        * Konto na FB
        * Punkty doswiadczenia na StackOverflow
    * Czlowiek nowoczesny ma metodę **programuj**, która zwiększa punkty doświadczenia 
    * Stworz property **czy_tamagotchi**
    * Stworz property **czy_aktywny_w_scoial_mediach**

```python
class Human ():
    def __init__(self, imie,nazwisko, wiek, czy_najedzony, czy_wyspany, czy_spragniony):
        self.imie = imie
        self.nazwisko = nazwisko
        self.wiek = wiek
        self.czy_najedzony = czy_najedzony
        self.czy_wyspany = czy_wyspany
        self.czy_spragniony = czy_spragniony
        
    @property
    def jesc (self):
        print ("Teraz jesz {}".format(self.imie))

    @property
    def pic (self):
        print ("Teraz pijesz {}".format(self.imie))
        
    @property
    def spac (self):
        print ("Teraz spisz {}".format(self.imie))
```

```python
h1 = Human('jan','nowak',32, True, True, True)
```

```python
h2=Human('tomcio','paluch',99,False, True, False)
```

```python
h1.spac
```

```python
class Super_human(Human):
    def __init__(self,imie,nazwisko, wiek, czy_najedzony, czy_wyspany, czy_spragniony, czy_git, czy_fb, pkt):
        super().__init__(imie,nazwisko, wiek, czy_najedzony, czy_wyspany, czy_spragniony)
        self.czy_git = czy_git
        self.czy_fb = czy_fb
        self.pkt = pkt
        
    
    def programuj(self,points):
        print ("Miales pkt {}".format(self.pkt))
        self.pkt  = self.pkt + points
        print ("Teraz masz pkt {}".format(self.pkt))
        
    @property
    def czy_tamagotchi (self):
        if (self.czy_najedzony and self.czy_wyspany and self.czy_spragniony):
            print ("Tak {} jestes Tamogotchi".format(self.imie))
    
```

```python
sh1 = Super_human('jack','sparrow',22,True, False, False, False, False, 200)
```

```python
sh2 = Super_human('James','Bond',25,True, True, False, True, False, 300)
```

```python
sh2.programuj (30)
```

```python
sh2.czy_tamagotchi
```
